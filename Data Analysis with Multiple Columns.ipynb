```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("googleplaystore.csv")

impute  = SimpleImputer(missing_values = np.nan , strategy = 'mean')
impute.fit(df.iloc[ : , 2:3].values)
df.iloc [ : ,2 : 3] = impute.transform(df.iloc[ : , 2:3].values)

df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df
```

## How many free apps are there in ART_AND_DESIGN


```python
df = df.values
```

```python
df
```

```python
c = 0
for i in df:
    if(i[1] == 'ART_AND_DESIGN' and i[6] == 'Free'):
        c+=1
print("There are total", c , "free apps from ART_AND_DESIGN")
```

# How many apps in ART_AND_DESIGN are there with the rating >4.5


```python
c = 0
for i in df:
    if(i[1] == 'ART_AND_DESIGN' and i[2] >= 4.5):
        c+=1
print("There are total", c , "apps from ART_AND_DESIGN with the rating more than 4.5")
```

## How many free, family apps are there in the with rating more than 4.0


```python
c = 0
for i in df:
    if(i[1] == 'FAMILY' and i[6] == 'Free' and i[2]>= 4.0):
        c += 1
print("There are total", c ,"apps.")
```

## List all the family apps that are free and with 5.0 rating

```python
print ("*"*90)
print("These are the apps that belongs to Family Category, are free and have a rating more than 5")
print ("*"*90)
for i in df:
    if(i[1] == 'FAMILY' and i[2]>=5.0 and i[6]=='Free'):
        print(i[0],i)
print("__"*50)
    
```
